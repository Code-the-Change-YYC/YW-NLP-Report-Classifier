## Incident Types

Check for new incident types.

````python
new = set([*new_df[ColName.INC_T1], *new_df[ColName.INC_T2]])
old = set([*old_df[ColName.INC_T1], *old_df[ColName.INC_T2]])
print(old - new)
print(new - old)
````

One entirely new type added: 'Refusal to isolate for covid-19'.

Examples of known types which previously lacked data added.

Calculate additional incident types examples provided in the new data.

````python
def get_inc_counts(colname: ColName):
    old = old_df[colname].value_counts()
    new = new_df[colname].value_counts()
    old = old.reindex(old.index.union(new.index)).fillna(0)
    return old, new

old1, new1 = get_inc_counts(ColName.INC_T1)
print((new1 - old1))
````

````python
old2, new2 = get_inc_counts(ColName.INC_T2)
print((new2 - old2))
````
